#### Here we are creating gradient descent to find better coefficient of determination and thus using this C.O.D to predict the value of Y_test using linear regression of n variable

```python
## Importing the Libraries

import numpy as np
import pandas as pd
from sklearn import preprocessing
```

```python
# Loading the training dataset

train_data = np.genfromtxt("train1.csv", delimiter = ",")

test_data = np.genfromtxt("test1.csv", delimiter = ",")


x_train = train_data[:, :-1]
y_train = train_data[:, -1]

train_df = pd.DataFrame(x_train)



col_len = len(train_df.values[0])

for i in range(col_len):
    train_df[col_len +i] = train_df[i] * train_df[i]
    
# print(train_df)

x_train = train_df.values

x_train_scale = x_train



## Test Data derived feature

test_df = pd.DataFrame(test_data)



col_len = len(test_df.values[0])

for i in range(col_len):
    test_df[col_len +i] = test_df[i] * test_df[i]
    
print(test_df)

x_test = test_df.values

x_test_scale = x_test

## Adding the X(n+1) column for finding the Intercept 
c = np.ones(x_train.shape[0]).reshape(-1, 1)
x_train = np.append(x_train, c, axis = 1)
```

```python
## StepGradient funcn

def stepGradient(x, y, lr, m_arr):
    slope = 0
    m = x.shape[0]
    for i in range(m):
        slope += (-2 / m) * (y[i] -(m_arr * x[i]).sum()) * x[i]
    
    m_arr = m_arr - lr * slope
    return m_arr
```

```python
## Defining the cost funcn

def cost(x, y, m_arr):
    m = x.shape[0]
    cost =0
    for i in range(m):
        cost += ((y[i] - (m_arr * x[i]).sum())**2)
    return cost / m
```

```python
## Gradient Descent funcn

def gradientDescent(x, y, lr, iteration):
    m_arr = np.zeros(x.shape[1])
    for i in range(iteration):
        m_arr = stepGradient(x, y, lr, m_arr)
        print("Learing Rate :- ", lr , " iteration :- ", i , " Cost :- ", cost(x, y, m_arr))
    return m_arr
```

```python
## Fit funcn or run() funcn

def fit(x, y):
    learning_rate = [0.001, 0.005, 0.01]
    iteration = 1000
    for i in range(len(learning_rate)):
        m_arr = gradientDescent(x, y, learning_rate[i], iteration)
        print("----------------------GD----------------------")
        print(m_arr)
    return m_arr
```

```python
## Getting coefficient of determination for training dataset without scaling 
m_arr0 = fit(x_train, y_train)


## I have print the learning rate, iteration & cost of all the iteration and learn. Rate
## Learning Rate = 0.1 Gives the best convergence and iteration it took till 350 
## For LR = 0.1 & iteration = 350 the cost diff. for last 2 iteration comes out to be 10^(-5)
```

```python
## Predict funcn for Predicting the o/p

def predict(x, m_arr):
    n = x.shape[0]
    y_p = np.zeros(n)
    for i in range(n):
        y_p[i] = (x[i] * m_arr).sum()
    return y_p
    
    
```

```python
## Funcn to calculate the score

def score(y_true, y_pred):
    u = ((y_true - y_pred)**2).sum()
    v = ((y_true - y_true.mean())**2).sum()
    return (1 - (u / v))
```

```python
## Feature Scaling for train as well as test data

scaler = preprocessing.StandardScaler()
scaler.fit(x_train_scale)

## Scaling the range of the training data
x_train_scaled = scaler.transform(x_train_scale)
x_train_scaled = np.append(x_train_scaled, np.ones(x_train_scaled.shape[0]).reshape(-1, 1), axis = 1)

## Scaling the range of the test data
x_test_scaled = scaler.transform(x_test)
x_test_scaled = np.append(x_test_scaled, np.ones(x_test_scaled.shape[0]).reshape(-1, 1), axis = 1)



x_train_scaled.shape, x_test_scaled.shape
```

```python
## Fitting the scaled train Data

m_arr1 = fit(x_train_scaled, y_train)

## I have print the learning rate, iteration & cost of all the iteration and learn. Rate
## Learning Rate = 0.1 Gives the best convergence and iteration it took till 350 
## For LR = 0.1 & iteration = 350 the cost diff. for last 2 iteration comes out to be 10^(-5)
```

```python
## Predicting the training values (Unscaled Training data)
y_tr_pred = predict(x_train, m_arr0)

## Predicting the training values (scaled Training data)
y_tr_pred_sc = predict(x_train_scaled, m_arr1)

## Predicting the testing values (scaled testing data)
y_tst_pred_sc = predict(x_test_scaled, m_arr1)

## Score of the unscaled features
print("Using unscaled features :-> ", score(y_train, y_tr_pred))

## Score of the scaled features
print("Using scaled features :-> ", score(y_train, y_tr_pred_sc))
```

```python
## Creating a csv file for submission
## Removing the exponent part also

np.savetxt('prediction2.csv', y_tst_pred_sc, delimiter = ",", fmt='%.14f')
```
